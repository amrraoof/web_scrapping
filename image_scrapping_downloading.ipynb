```python
"""This notebook is an implementaion of "How to scrap and download an image from a given website which you know the specific class of that image" """
from bs4 import BeautifulSoup
import requests
import urllib.request

page = requests.get("https://unsplash.com/s/photos/scrap").text
soup = BeautifulSoup(page,"lxml")

print(soup.title)
image = soup.find("img", class_ = "tB6UZ a5VGX" )
```

```python
#print(soup.prettify())
image_link = image.attrs["src"]
print(image_link)
```

```python
image = requests.get(image_link)
```

```python
print(image) # Ensure that the link was downloaded successfully
```

```python
# Now let's save the image 
with open("scraped_image.png" , "wb") as new_image:
    for chunck in image.iter_content(chunk_size=1024):
        new_image.write(chunck)
```
